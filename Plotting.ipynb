**3. Pascal's Triangle** *(10 points)*

Use a `for` loop and the binomial coefficients defined in Week 2's exercises to print out the first 20 lines of Pascal's triangle. The nth line of Pascal's triangle contains n + 1 numbers that are the binomial coefficients ${n\choose 0}$, ${n\choose 1}$, and so on up to ${n\choose n}$. Thus the first few lines are

`1` <br>
`1 1` <br>
`1 2 1` <br>
`1 3 3 1` <br>

Coding hint: Normally, when you call the `print` function, the output automatically goes to the next line.
If you want the next thing printed to be on the same line, you can supply the argument `end=' '` to `print`.
Then,  to go to the next line, simply write `print()`.  

```python
def fact(n):
    retVal = 1
    while n>0:
        retVal*=n
        n-=1
    return retVal

def binomial(n, k):
    #This is defined as n choose k
    denom = fact(k)*fact(n-k)
    numer = fact(n)
    return int(numer / denom)

def pascalTri(rowMax):
    for i in range(rowMax):
        print("")
        for j in range(i+1):
            print(binomial(i, j), end=' ')

pascalTri(20)

```

**9. Cartesian to Polar coordinates.**

The formulas for converting Cartesian coordinates $x,y$ to polar coordinates $r,\theta$ are
$$
r = \sqrt{x^2+y^2},\ \ \ \theta = \arctan(y/x).
$$

a. *(5 points)* Define a function using `def()` that takes $x,y$ as arguments and returns $r,\theta$.

b. *(5 points)* If you enter $x=0$, your program will give an error (why is this?).  Modify your code to exit when you enter $x=0$, as well as to return the angle in degrees.

```python
def radToDeg(rad):
    return rad*360/(2*pi)

def cartToPolar(x, y):
    r = float(sqrt(x*x + y*y))
    try:
        theta = np.arctan(y/x)
    except ZeroDivisionError:
        if y>0:
            theta = pi / 2
        else:
            theta = - pi/2
    return(r, radToDeg(theta))

cartToPolar(0, -5)
```

By convention $\arctan(..)$ always returns an angle in the interval $(\pi/2,\pi/2)$.

Check that this version of your program works for $x,y$ in any quadrant.

```python
def cartToPolar(x, y):
    r = float(sqrt(x*x + y*y))
    try:
        theta = np.arctan(y/x)
    except ZeroDivisionError:
        if y>0:
            theta = pi / 2
        else:
            theta = - pi/2
    if x>0:
        theta = theta
    if x<0:
        theta += pi
    return(r, float(radToDeg(theta)))

print(cartToPolar(-5, 10))

def cartToPolarWithArc(x, y):
    r = float(sqrt(x*x + y*y))
    theta = np.arctan2(y, x)
    return(r, float(radToDeg(theta)))

print(cartToPolarWithArc(-5, 10))
```
